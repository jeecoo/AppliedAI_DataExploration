```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_csv('paris_olympics2024\\paris olympics 2024 dataset\\medallists.csv')

# Check if 'medal_type' column exists
if 'medal_type' in df.columns:
    # Extract the 'medal_type' column
    medal_type = df['medal_type']
    
    # Count occurrences of each type of medal
    medal_counts = medal_type.value_counts()
    print(f'{df['medal_type'].describe()}\n')
    
    print(medal_counts)

    # Visualizing the 'medal_type' column using Seaborn
    plt.figure(figsize=(10, 6))
    
    # Bar plot with hue assigned to 'x' variable and legend disabled
    sns.barplot(x=medal_counts.index, y=medal_counts.values, hue=medal_counts.index, palette='viridis', legend=False)
    
    # Adding title and labels
    plt.title('Distribution of Medals', fontsize=16)
    plt.xlabel('Medal Type', fontsize=14)
    plt.ylabel('Count', fontsize=14)

    # Show the plot
    plt.show()

else:
    print("The 'medal_type' column is not found in the CSV file.")
```

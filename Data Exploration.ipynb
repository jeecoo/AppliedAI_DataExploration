## Anscombe's Quartet

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load Anscombe's quartet dataset
anscombe = sns.load_dataset("anscombe")

# Show the dataset's size and shape
print("Size:",anscombe.size)
print("Shape:",anscombe.shape)
```

```python
# Show entire dataset
anscombe
```

```python
# Create a grid of plots for each dataset in the quartet
g = sns.FacetGrid(anscombe, col="dataset", col_wrap=2, height=4)

# Map the points to the grid and add a linear regression line (without confidence intervals)
g.map(sns.regplot, "x", "y", ci=None)

# Add titles
g.set_titles("Dataset {col_name}")

# Add labels to the axes
g.set_axis_labels("X", "Y")

# Show the plot
plt.tight_layout()
plt.show()
```

# Iris Dataset

```python
# Load the Iris dataset
iris = sns.load_dataset("iris")

# Melt the dataset so that we can plot all the features together
iris_melted = pd.melt(iris, id_vars=["species"], value_vars=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                      var_name="Feature", value_name="Value")

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a box plot for all features
plt.figure(figsize=(10, 6))
sns.boxplot(x="Value", y="Feature", data=iris_melted)

# Add title and labels
plt.title("Box plot of all Iris features")
plt.ylabel("Features")
plt.xlabel("Value (cm)")

# Show the plot
plt.tight_layout()
plt.show()
```

```python

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a box plot for the Iris dataset
plt.figure(figsize=(10, 6))
sns.boxplot(y="species", x="sepal_length", data=iris)

# Add labels and title
plt.title("Box plot of Sepal Length by Species")
plt.ylabel("Species")
plt.xlabel("Sepal Length (cm)")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a box plot for each feature by species
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(y="species", x=feature, data=iris)
    plt.title(f"Box plot of {feature.replace('_', ' ').capitalize()} by Species")
    plt.ylabel("Species")
    plt.xlabel(f"{feature.replace('_', ' ').capitalize()} (cm)")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Create a pair plot with regression lines
sns.pairplot(iris, kind="reg", diag_kind="kde", plot_kws={'ci': None})

# Show the plot
plt.tight_layout()
plt.show()
```

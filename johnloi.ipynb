```python
import pandas as pn
import numpy as np
import matplotlib.pyplot as plt

df = pn.read_csv(r'paris_olympics2024\paris olympics 2024 dataset\schedules.csv')

discipline_event_count = df['discipline'].value_counts()

plt.figure(figsize=(10, 6))
plt.hist(discipline_event_count, bins=10, color='skyblue', edgecolor='black') 
plt.title('Distribution of Number of Events per Discipline')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
